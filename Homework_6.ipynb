```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report


np.random.seed(42)

n = 50
train_data = {
    "monthly": np.random.randint(500, 3000, n),
    "credit": np.random.randint(450, 800, n),
    "employment_status": np.random.randint(0, 2, n),
    "existing_loans": np.random.randint(0, 4, n),
    "age": np.random.randint(20, 55, n)
}

train_data = {
    "monthly": [800, 1200, 1500, 2000, 500, 900, 1800, 2200, 1300, 2500],#it salaray
    "credit": [520, 600, 650, 720, 480, 590, 700, 740, 630, 780],
    "employment_status": [0, 1, 1, 1, 0, 1, 1, 1, 1, 1],  # 0=unemployed, 1=employ
    "existing_loans": [2, 1, 1, 0, 3, 2, 0, 0, 1, 0],#,1=have loans ,2=have loans ,3=have loans,0=no existing loans
    "age": [22, 28, 35, 41, 20, 30, 38, 45, 27, 50],
    "approved_score": [40, 60, 65, 80, 30, 55, 78, 85, 62, 90]
}
df1 = pd.DataFrame(train_data)

df1["approved"] = (df1["approved_score"] >= 60).astype(int)

monthly = float(input("Please input monthly income: "))
credit = int(input("Please input credit score: "))

emp_raw = input("Please input employment status (1=employed, 0=unemployed): ").strip()
employment_status = int(emp_raw)

existing_loans = int(input("Please input existing loans: "))

while True:
    try:
        age = int(input("Please input age: "))
        if age < 1:
            print("Age must be greater than 0")
            continue
        break
    except ValueError:
        print("Invalid input. Please enter a number")

df_new = pd.DataFrame([{
    "monthly": monthly,
    "credit": credit,
    "employment_status": employment_status,
    "existing_loans": existing_loans,
    "age": age
}])

X1 = df1[["monthly", "credit", "employment_status", "existing_loans", "age"]]
y1 = df1["approved"]

X_train1, X_test1, y_train1, y_test1 = train_test_split(
    X1, y1, test_size=0.2, random_state=42, stratify=y1
)
model1 = LogisticRegression(max_iter=1000)
model1.fit(X_train1, y_train1)

y_pred1 = model1.predict(X_test1)
print("Accuracy:", accuracy_score(y_test1, y_pred1))
print("Confusion Matrix:\n", confusion_matrix(y_test1, y_pred1))
print("Report:\n", classification_report(y_test1, y_pred1))

prediction1 = model1.predict(df_new[X1.columns])[0]
prob1 = model1.predict_proba(df_new[X1.columns])[0][1]

print("\nPrediction:", "Approved" if prediction1 == 1 else "Rejected")
print("Approval probability:", round(prob1 * 100, 2), "%")

if prediction1 == 0:
    print("\recommend to Improve Approval Chances:")

    if credit < 600:
        print("- Improve your credit score (target 650+)")

    if existing_loans > 1:
        print("- Reduce your existing loans")

    if monthly < 1000:
        print("- Increase your monthly income")

    if employment_status == 0:
        print("- Stable employment increases approval chances")

    if age < 21:
        print("- Very young applicants may have lower approval rates")
else:
    print("\nCongratulations! You meet the basic approval conditions.")
```
