```python
#============Assignment 1 ===========================
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
monthly_sales = [1200, 1500, 1700, 1600, 1800, 2000]

products = ["Product A", "Product B", "Product C", "Product D"]
product_sales = [500, 700, 600, 900]

plt.figure()
plt.plot(months, monthly_sales, label="Monthly Sales")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.legend()
plt.show()


plt.bar(products, product_sales, label="Product Sales")
plt.title("Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales ($)")
plt.legend()
plt.show()
```

```python
#============Assignment 2 ===========================
students = ["Alice", "Bob", "Charlie", "David", "Eva"]
scores = [85, 92, 78, 88, 95]

highest_score = max(scores)
highest_index = scores.index(highest_score)


image = plt.figure()
bars = plt.bar(students, scores)


plt.title("Student Scores Analysis")
plt.xlabel("Students")
plt.ylabel("Scores")


bars[highest_index].set_linewidth(3)
bars[highest_index].set_hatch("///")

fig = plt.gcf()
fig.savefig("./homework/assignment/storage/assignment2_plot.png", dpi=300)

plt.show()

```

```python
#==================Assignment 3 ===========================
import matplotlib.pyplot as plt
import numpy as np

years = np.array([2015, 2016, 2017, 2018, 2019])
branches = ["IT", "ECE", "CSE"]

passed = {
    "IT":  np.array([12, 30,  1,  8, 22]),
    "ECE": np.array([28,  6, 16,  5, 10]),
    "CSE": np.array([29,  3, 24, 25, 17]),
}

\
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(9, 11))

\
x = np.arange(len(years))
width = 0.25

axes[0].bar(x - width, passed["IT"],  width, label="IT",  color="red")
axes[0].bar(x,         passed["ECE"], width, label="ECE", color="green")
axes[0].bar(x + width, passed["CSE"], width, label="CSE", color="blue")

axes[0].set_title("Students Passed by Branch (Grouped Bar Chart)")
axes[0].set_ylabel("Students Passed")
axes[0].set_xticks(x)
axes[0].set_xticklabels(years)
axes[0].legend()


for b in branches:
    axes[1].plot(years, passed[b], marker="o", label=b)
axes[1].set_title("Trend Over Years (Line Chart)")
axes[1].set_ylabel("Students Passed")
axes[1].set_xlabel("Year")
axes[1].legend()


matrix = np.vstack([passed[b] for b in branches]) 
im = axes[2].imshow(matrix, aspect="auto")
axes[2].set_title("Heatmap of Students Passed (Branch x Year)")
axes[2].set_xlabel("Year")
axes[2].set_ylabel("Branch")
axes[2].set_xticks(np.arange(len(years)))
axes[2].set_xticklabels(years)
axes[2].set_yticks(np.arange(len(branches)))
axes[2].set_yticklabels(branches)
fig.colorbar(im, ax=axes[2], label="Students Passed")

fig.suptitle("Assignment 3  Mini Visualization Project (Exam Results)", y=0.995)
fig.tight_layout()

```
