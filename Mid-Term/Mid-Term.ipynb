```python
#=======================================Task 2============================================
marks = []

for i in range(5):
    m = float(input(f"Enter mark for student {i+1}: "))
    marks.append(m)

average = sum(marks) / len(marks)

print("Marks:", marks)
print("Average:", average)

if average >= 50:
    print("Pass")
else:
    print("Fail")
```

```python
#=======================================Task 3============================================
import numpy as np

A = np.array([[1, 2, 3],
              [0, 4, 5],
              [1, 0, 6]])

print("Matrix A:\n", A)


AT = A.T
print("\nTranspose:\n", AT)


detA = np.linalg.det(A)
print("\nDeterminant:", detA)


if detA != 0:
    invA = np.linalg.inv(A)
    print("\nInverse:\n", invA)
else:
    print("\nInverse does not exist (determinant is 0).")
```

```python
#=======================Task 4============================================
import pandas as pd

df = pd.DataFrame({
    "Name": ["Ali", "Sara", "John", "Mina", "David"],
    "Age":  [18, 19, 18, 20, 19],
    "Score":[65, 82, 74, 49, 91]
})

print("DataFrame:\n", df)


print("\nFirst 3 rows:\n", df.head(3))


print("\nScore > 70:\n", df[df["Score"] > 70])


df["Status"] = df["Score"].apply(lambda x: "Pass" if x >= 50 else "Fail")
print("\nWith Status column:\n", df)
```

```python
#Part II
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#Generate 20 random Number
random_numbers = np.random.randint(0, 100, 20)
print(random_numbers)

#Covert Pandas to DataFrame
df = pd.DataFrame(random_numbers, columns=["Numbers"])
print("Data Frame",df)


#Calculate Mean Median and Standard Deviation
mean = df["Numbers"].mean()
median = df["Numbers"].median()
std = df["Numbers"].std()

print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std)

#Plot Histogram
plt.hist(df["Numbers"], bins=10)

plt.title("Histogram of Random Numbers")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()

#Explain distribution

# Distribution = "Distribution គឺជាការពិពណ៌នាអំពីរបៀបដែលទិន្នន័យត្រូវបានចែកចាយ។ វាបង្ហាញពីរបៀបដែលតម្លៃក្នុងទិន្នន័យត្រូវបានចែកចាយ និងអាចបង្ហាញពីលំនាំ (patterns) ឬការប្រមូលផ្តុំ (clusters) ក្នុងទិន្នន័យ។"
```
