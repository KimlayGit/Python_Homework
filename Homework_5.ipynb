```python
#========Practice 2========
Subjects = ["Math", "Science","English"]
Marks = [80, 75, 90]
plt.bar(Subjects, Marks,)
plt.title("Marks in Different Subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
#========Practice 3========
data = np.random.rand(100)
plt.hist(data, bins=15, edgecolor='black')
plt.title("Histogram of Random Data")
plt.show()
```

```python
#========Practice 4========
plt.pie([15.6, 23.2, 12.9, 8.2, 8.2, 32], labels=["Samsung", "Apple", "Xiaomi", "Transsion","Vivo", "Others"], autopct='%1.1f%%')
plt.title("Global Smartphone Market Share in 2025")
plt.show()
```

```python
#========Practice 5========
x = [1, 2, 3, 4, 5]
y_line = [10, 15, 7, 12, 18]
y_bar = [5, 8, 6, 9, 4]

plt.figure()

plt.bar(x, y_bar)

plt.plot(x, y_line)

plt.title("Combined Line and Bar Chart")
plt.xlabel("X Values")
plt.ylabel("Y Values")

plt.show()
```
